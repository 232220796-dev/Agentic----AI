```python
%%writefile faq.txt

Password reset:
Users can reset their password from account settings.

Billing:
Refund requests are processed within 7 business days.

Technical:
Application errors can be solved by restarting and updating the application.

Account:
Account verification requires email confirmation.
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma


loader = TextLoader("faq.txt")

docs = loader.load()


splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50
)

chunks = splitter.split_documents(docs)


embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)


vectorstore = Chroma.from_documents(
    chunks,
    embeddings
)


retriever = vectorstore.as_retriever()


print("RAG Ready")
```

```python
from transformers import pipeline


llm = pipeline(
    "text-generation",
    model="google/flan-t5-small",
    max_new_tokens=100
)
```

```python
def search_kb(question):

    docs = retriever.invoke(question)

    return "\n".join(
        [d.page_content for d in docs]
    )
```

```python
def technical_agent(state):

    info = search_kb(state["message"])

    answer = llm(
        f"Answer customer using information:\n{info}\nQuestion:{state['message']}"
    )[0]["generated_text"]

    state["response"] = answer

    return state



def billing_agent(state):

    info = search_kb(state["message"])

    answer = llm(
        f"Answer billing question:\n{info}\nQuestion:{state['message']}"
    )[0]["generated_text"]

    state["response"] = answer

    return state
```

```python
def supervisor(state):

    msg = state["message"].lower()

    if "refund" in msg or "payment" in msg:
        state["next"]="billing"

    elif "error" in msg or "bug" in msg:
        state["next"]="technical"

    else:
        state["next"]="human"

    return state
```

```python
from langgraph.types import interrupt


def human(state):

    answer = interrupt(
        "Human support required"
    )

    state["response"]=answer

    return state
```

```python
from typing import TypedDict
from langgraph.graph import StateGraph, END


class State(TypedDict):
    message:str
    next:str
    response:str


graph = StateGraph(State)


graph.add_node("supervisor", supervisor)
graph.add_node("billing", billing_agent)
graph.add_node("technical", technical_agent)
graph.add_node("human", human)


graph.set_entry_point("supervisor")


graph.add_conditional_edges(
    "supervisor",
    lambda x:x["next"],
    {
        "billing":"billing",
        "technical":"technical",
        "human":"human"
    }
)


graph.add_edge("billing", END)
graph.add_edge("technical", END)
graph.add_edge("human", END)


app = graph.compile()


print("Agent System Ready")
```

```python
result = app.invoke(
    {
        "message":"I need a refund"
    }
)

print(result["response"])
```

```python
from langgraph.checkpoint.memory import MemorySaver

memory = MemorySaver()

app = graph.compile(
    checkpointer=memory
)
```

```python
config = {
    "configurable": {
        "thread_id": "customer_1"
    }
}

result = app.invoke(
    {
        "message":"I need a refund"
    },
    config
)

print(result["response"])
```
